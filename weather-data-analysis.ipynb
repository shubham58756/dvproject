```python
import pandas as pd

# Replace 'your_file_path' with the actual path to your CSV file
file_path = '/kaggle/input/weather-data/Weather Data.csv'
data = pd.read_csv(file_path)
import matplotlib.pyplot as plt

# Line graph to visualize temperature trends
plt.figure(figsize=(10, 6))
plt.plot(data['Date/Time'], data['Temp_C'], label='Temperature (°C)')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Trends Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file_path' with the actual path to your CSV file
file_path = '/kaggle/input/weather-data/Weather Data.csv'
data = pd.read_csv(file_path)

# Histogram to visualize temperature distribution
plt.figure(figsize=(8, 6))
plt.hist(data['Temp_C'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Distribution of Temperature (°C)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file_path' with the actual path to your CSV file
file_path = '/kaggle/input/weather-data/Weather Data.csv'
data = pd.read_csv(file_path)

# Scatter plot to explore the relationship between temperature and wind speed
plt.figure(figsize=(8, 6))
plt.scatter(data['Temp_C'], data['Wind Speed_km/h'], alpha=0.5)
plt.xlabel('Temperature (°C)')
plt.ylabel('Wind Speed (km/h)')
plt.title('Scatter Plot: Temperature vs. Wind Speed')
plt.show()
```
